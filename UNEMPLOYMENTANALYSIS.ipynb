```python
#plotting correlation heatmap

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8,6))

#Set tick parameters
ax = plt.gca()
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='w')
ax = sns.heatmap(data.corr(), annot=True,linewidth=3)
plt.show()
```

```python
#plotting histplot
data.columns=['Region','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation', 'Area']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

Analyzing Data By Bar Graph

```python
fg = px.bar(df,x='Region',y='Estimated Unemployment Rate',color='Region',
            title='Unemploymeny Rate (Region Wise) by Bar Graph',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

```python
fg = px.bar(df,x='Area',y='Estimated Unemployment Rate',color='Area',
            title='Unemploymeny Rate (Area Wise) by Bar Graph',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

Analyzing Data By Histogram

```python
fg = px.histogram(df,x='Region',y='Estimated Unemployment Rate',color='Region',
                  title='Unemploymeny Rate (Statewise) by Histogram',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

Analyzing Data By Scatterplot

```python
fg = px.scatter(df,x='Region',y='Estimated Unemployment Rate',color='Region',
                title='Unemploymeny Rate (Statewise) by Scatter Plot',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```
